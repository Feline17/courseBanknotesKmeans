```python
# imports in one place, research if it can lead to errors
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#import dataset

data = pd.read_csv("Banknote-authentication-dataset-.csv")
# print(data)
```

```python
# mean of both vars
data.mean() 
```

```python
# other formula, also gives mean

np.mean(data)
```

```python
# standard deviation of vars
np.std(data)
```

```python
# Shows max and min of both vars

print((data.min()), 
      (data.max()))
```

```python
# make columns from dataset into variables to use in graph

V1 = data["V1"]
V2 = data["V2"]

# x-axis and y-axis labels
plt.xlabel("V1")
plt.ylabel("V2")

# plot a scatter graph
plt.scatter(V1, V2)
```

```python

# Yes, this dataset seems suitable for the K-means clustering task 
# because most of the data points seem to be heavily concentrated in some areas
# hinting at the possibilty of formation around clusters.
# For example, heavy = 0,0; light = 6,10
```

```python
# Step 1: run K-means on the given dataset

#Step 2: visualise the results.

#Step 3: re-run K-means several times and visualise the results.

#Step 4: compare the results: is the K-means algorithm stable?

#Step 5: describe your results.
```

```python
# covert 2 1D lists into a 2D one which will be used for KMeans

V1_V2 = np.column_stack ((V1, V2))
print(V1_V2)
```

```python
#KMeans formula!!!

x = 0 
while x < 10:
    km_res = KMeans(n_clusters=3).fit(V1_V2) # how many clusters and which 2D array
    clusters = km_res.cluster_centers_ #what are the cluster coordinates?
    print(clusters)
    x += 1
```

```python
#plot graph and clusters
plt.scatter(V1,V2)
plt.scatter(clusters[:,0], clusters[:,1], s = 100) #need 1D arrays like before # also, size is 100
```

```python
# add them all up! perfect graph

# x-axis and y-axis labels
plt.xlabel("V1")
plt.ylabel("V2")

# plot a scatter graph
plt.scatter(V1, V2)

plt.scatter(clusters[:,0], clusters[:,1], s = 100) #need 1D arrays like before # also, size is 100
```
