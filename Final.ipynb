```python
# imports in one place, research if it can lead to errors
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#import dataset

data = pd.read_csv("Banknote-authentication-dataset-.csv")
# print(data)
```

```python
# make columns from dataset into variables to use in graph

V1 = data["V1"]
V2 = data["V2"]
```

```python
# covert 2 1D lists into a 2D one which will be used for KMeans

V1_V2 = np.column_stack ((V1, V2))
print(V1_V2)
```

```python
#KMeans formula!!!

x = 0 
while x < 10:
    km_res = KMeans(n_clusters=2).fit(V1_V2) # how many clusters and which 2D array
    clusters = km_res.cluster_centers_ #what are the cluster coordinates?
    #print(clusters)
    x += 1
```

```python
# graph

# x-axis and y-axis labels
plt.xlabel("V1")
plt.ylabel("V2")

# plot a scatter graph
plt.scatter(V1, V2)

plt.scatter(clusters[:,0], clusters[:,1], s = 100) #need 1D arrays like before # also, size is 100
```

```python
#import dataset

data = pd.read_csv("orignalbanknote5varopenml.csv")

# make columns from dataset into variables to use in graph

V1 = data["V1"]
V2 = data["V2"]

# x-axis and y-axis labels
plt.xlabel("V1")
plt.ylabel("V2")

plt.scatter(V1, V2) 
```

```python
# filter

genuine = data[data["Class"] == 1]
forged = data[data["Class"] == 2]

plt.scatter(V1, V2)

# mark all genuine points

for k,row in genuine.iterrows():
    plt.text(row["V1"],
            row["V2"],
            row["Class"])
```

```python
# all genuine points should be green 

for k,row in genuine.iterrows():
    plt.scatter(row["V1"], row["V2"], c = "green", alpha = 0.25)

# all forged points should be red 

for k,row in forged.iterrows():
    plt.scatter(row["V1"], row["V2"], c = "red", alpha = 0.15)
```

```python
# code from peer in forum who was able to

V1_V2 = np.column_stack ((V1, V2))

km_res = KMeans (n_clusters= 2).fit(V1_V2)
y_kmeans = km_res.predict(V1_V2)

plt.scatter(V1, V2, c = y_kmeans, s = 50, alpha = 0.5)
clusters = km_res.cluster_centers_ #what are the cluster coordinates?
    #print(clusters)
    
plt.scatter(clusters[:,0], clusters[:,1], c="black", s = 100)
```
